### Analyzing Complexity

Finding out the computation time of 3 very simple problems and 3 sorting methods

```python
from matplotlib.pyplot import plot
from matplotlib import pyplot
import numpy as np
import timeit
from functools import partial
import random
from sklearn.utils import shuffle
```

```python
def sumbetween_n(n):
    lst = np.arange(n)
    result = 0
    for i in lst:
        result += i
    return result

def between_n_simple(n):
    return (n*(n+1))/2

def simple_sum(n):
    return sum(np.arange(n))

def plotTC(fn,nMin,nMax, nInc, nTest):
    x = []
    y = []
    for i in range(nMin,nMax, nInc):
        N = i
        testNTimer = timeit.Timer(partial(fn,N))
        t = testNTimer.timeit(number = nTest)
        x.append(i)
        y.append(t)
    p1 = plot(x,y,'-')
def main():
    print('Analyze algo...')
    plotTC(sumbetween_n,10,1000,10,10)
    plotTC(between_n_simple,10,1000,10,10)
    plotTC(simple_sum,10,1000,10,10)
    pyplot.show()
```

```python
if __name__ == '__main__':
    main()
```

```python
def bubble_sort(arr):
    def swap(i, j):
        arr[i],arr[j] = arr[j],arr[i]
    n = len(arr)
    swapped = True
    
    x = -1
    while swapped:
        swapped = False
        x = x+1
        for i in range(1, n-x):
            if arr[i-1] > arr[i]:
                swap(i - 1, i)
                swapped = True
    return arr

def selection_sort(arr):
    for i in range(len(arr)):
        minimum = 1
        
        for j in range(i+1, len(arr)):
            if arr[j] < arr[minimum]:
                minimum = j
            arr[minimum],arr[i] = arr[i],arr[minimum]
    return arr

def insetion_sort(arr,simulation = False):
    for i in range(len(arr)):
        cursor = arr[i]
        pos = i
        
        while pos > 0 and arr[pos-1] > cursor:
            arr[pos] = arr[pos-1]
            pos = pos - 1
            arr[pos] = cursor
    return arr
```

```python
def plotSO(fn,nMin,nMax, nInc, nTest):
    x = []
    y = []
    
    for i in range(nMin,nMax, nInc):
        N = np.arange(i)
        M = shuffle(N)
        testNTimer = timeit.Timer(partial(fn,M))
        t = testNTimer.timeit(number = nTest)
        x.append(i)
        y.append(t)
    p1 = plot(x,y,'-')
```

```python
def main():
    print('Analyze algo...')
    plotSO(bubble_sort,10,1000,10,10)
    plotSO(selection_sort,10,1000,10,10)
    plotSO(insetion_sort,10,1000,10,10)
    pyplot.show()
```

```python
if __name__ =='__main__':
    main()
```
